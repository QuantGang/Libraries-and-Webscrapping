Con esta función se pueden descargar series de tiempo desde Alpha Vantage del tipo 'stock'(acciones) o 'cripto' (criptomoneda) 'ticker', según una frecuencia 'frecuency' que puede ser 'intraday', 'daily', 'weekly' o 'monthly'. Los parámetros para cada frecuencia son los siguientes:

1.   'intraday': se debe definir un intervalo de tiempo que puede ser '1min', '5min', '15min', '30min' or '60min' y un tamaño de datos que puede ser 'full' or'compact' el cual incluye los últimos 100 valores. No se pueden descargar datos por fechas.
Ejemplo sintaxis: 
data = odata(ticker='AAPL', interval='5min', outputsize='full', frecuency='intraday')
2.  'daily': No debe incluir 'interval'
Ejemplo sintaxis: 
data = odata(ticker='AAPL', outputsize='full', frecuency='intraday')

3.  'weekly' and 'monthly': solo incluyen 'ticker'
Ejemplo sintaxis: 
data = odata(ticker='AAPL')

Para las criptomonedas se debe definir un mercado específico asociado a una moneda, por ejemplo, market='COP', la lista completa se puede encontrar en https://www.alphavantage.co/documentation/ además no se pueden bajar series de tiempo para criptomonedas en frecuencia intradía.


```
def odata(type, ticker, interval = '30min', outputsize = 'full', frecuency = 'daily', market = 'USD'):
    API_KEY='F1SXLTFIY8B0W6P2'
    if type == 'stock':
      from alpha_vantage.timeseries import TimeSeries
      ts = TimeSeries(key=API_KEY,output_format='pandas', indexing_type='date')
      if frecuency == 'intraday':
        data = ts.get_intraday(symbol = ticker, interval = interval, outputsize = outputsize)
      elif  frecuency == 'daily':
        data = ts.get_daily(symbol = ticker, outputsize = outputsize)
      elif  frecuency == 'weekly':
        data = ts.get_weekly(symbol = ticker)
      elif  frecuency == 'monthly':
        data = ts.get_monthly(symbol = ticker)
      else: print('Datos no encontrados!!!')
      return data
    elif type == 'cripto':
      from alpha_vantage.cryptocurrencies import CryptoCurrencies
      cc = CryptoCurrencies(key=API_KEY, output_format='pandas')
      if frecuency == 'daily':
        data = cc.get_digital_currency_daily(symbol = ticker, market = market)
      elif  frecuency == 'weekly':
        data = cc.get_digital_currency_weekly(symbol = ticker, market = market)
      elif  frecuency == 'monthly':
        data = cc.get_digital_currency_monthly(symbol = ticker, market = market)
      else: print('Datos no encontrados!!!')
      return data
```

```
#Example
import pandas as pd
data = pd.DataFrame(odata('stock', ticker = 'CSCO', interval = '1min', outputsize = 'full', frecuency = 'intraday', market = 'USD'))
print(data[0][0])
```
