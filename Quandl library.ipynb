```
import pandas as pd
import quandl
import matplotlib.pyplot as plt
from datetime import date
```

A continuación creamos una función que obtenga la información de quandl con
  dos opciones posibles que son gratuitas: CHRIS (wiki) o EOD.
  Se crea una variable respuesta para cada base de datos.

```
# Obtenemos la información
def getdata(ticker,start,end=date.today()):
    response = input('The database is ')
    response = response.upper()
    if response == 'CHRIS':
        data = quandl.get('CHRIS/'+ticker, start_date=start, end_date = end)
        return data
    elif response == 'EOD':
        data = quandl.get('EOD/'+ticker, start_date=start, end_date=end)
        return data
    else:
        return 'No valid database'
```
